```python
import pandas as pd
import numpy as np
import scipy as sc
from scipy import constants
from scipy.signal import find_peaks
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit


plt.rcParams['text.usetex'] = True #LaTeX implementation
font1 = {'family':'sans-serif','color':'black','size':12}
```

### Aufgabe 1: 100% modulation index

#### FFT

```python
#Daten_einlesen
Daten = np.loadtxt('B28.CSV',delimiter=',', skiprows=1)
freq1_100 = Daten[:,0]

Resolution = 2000000/len(freq1_100)
freq_1_100_err = Resolution/2

FFT1_100 = Daten[:,1] 

FFT1_100_err = np.full(len(FFT1_100), 0.17)

plt.plot(freq1_100,FFT1_100,marker='x',linestyle = 'none')
```

```python
# Gesamte Peaks finden

selected_peaks,_ = find_peaks(FFT1_100,height=-40)



FFT1_100_height = FFT1_100[selected_peaks]
freq1_100_position = freq1_100[selected_peaks]



print("Gefundene Peaks bei:", freq1_100_position)
print("Peak-Höhen:", FFT1_100_height)

hdistance_100_1 = FFT1_100_height[1] - FFT1_100_height[0] 
hdistance_100_2 = FFT1_100_height[1] - FFT1_100_height[2] 

print(f"Abstand peak 1 und 2: {hdistance_100_1:.2f}(17)")
print(f"Abstand peak 2 und 3: {hdistance_100_2:.2f}(17)")


modulation_index_100_1 = 2 * FFT1_100_height[0] / FFT1_100_height[1]
modulation_index_100_2 = 2 * FFT1_100_height[2] / FFT1_100_height[1]

print(f"Modulation index (lower sideband): {modulation_index_100_1:.2f}")
print(f"Modulation index (upper sideband): {modulation_index_100_2:.2f}")


# Plot
plt.plot(freq1_100, FFT1_100, label="Signal")
plt.plot(freq1_100_position, FFT1_100_height, "rx", label="Gefundene Peaks")
plt.legend()
plt.show()
```

```python
#Plot
plt.figure(dpi=200)
plt.errorbar(freq1_100, FFT1_100, yerr=FFT1_100_err, xerr=freq_1_100_err, fmt='o', markersize=5, capsize=3, capthick=1,
             elinewidth=1.2, ecolor='black', markerfacecolor='tab:blue',
             markeredgecolor='black', label=r'Measurement with errorbar')
plt.ylabel(rf"FFT in dBm", fontdict=font1)
plt.xlabel(rf"Frequency in Hz", fontdict=font1)
plt.legend(loc='upper center', bbox_to_anchor=(0.5, -0.11),
           fancybox=True, shadow=True, ncol=1)

plt.show()
```
